```python
dfImpiegoRisorse.rename(columns={'Budget/cons': 'budget'}, inplace=True)
dfImpiegoRisorse = sqldf(
"SELECT * FROM dfImpiegoRisorse WHERE budgetConsuntivo = 'CONSUNTIVO' ")

print("Numero di righe: "+str(dfImpiegoRisorse.shape[0]))

dfImpiegoRisorse.head()
```
