```python
A = sqldf(
    "SELECT * FROM dfCostoOrarioConsuntivo O JOIN dfImpiegoRisorse C ON O.Risorsa = C.Risorsa AND O.AreaProduzione = C.NrAreaProduzione WHERE C.budgetConsuntivo = 'CONSUNTIVO' OR 'Consuntivo'")
print("Numero di righe = "+str(A.shape[0]))

del A['Unnamed: 0']
del A['Risorsa']

A.head()
```

```python
B= sqldf(
    "SELECT * FROM dfCostoOrarioBudget O JOIN dfImpiegoRisorse C ON O.Risorsa = C.Risorsa AND O.AreaProduzione = C.NrAreaProduzione WHERE C.budgetConsuntivo ='BUDGET'")
print("Numero di colonne: "+str(B.shape[0]))

del B['Unnamed: 0']
del B['Risorsa']


B.head()
```

```python
outer_join_df = pd.merge(A, B, on='NrArticolo', how='outer')
print("Numero di colonne con outer join = "+str(outer_join_df.shape[0]))

normale = sqldf("SELECT * FROM A A JOIN B B ON A.NrArticolo = B.NrArticolo")
print("Numero di colonne con  sqldf = "+str(normale.shape[0]))
```
